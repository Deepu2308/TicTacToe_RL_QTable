# Parse Logs

```python
with open( '../src/logs/sarsa.log','r') as f:
    log_txt = f.read()
```

```python
lines = log_txt.split('\n')
lines
meta = []
results = []

for i in lines:
    if 'experiment' in i.lower():
        meta.append(i)
    else:
        results.append(i)
```

```python
record_list = []
for record in results:    
    d = {}
    for col in record.split('|'):
        key,value = [i.strip() for i in col.split(":")]
        d[key] = value
    record_list.append(d)
```

# Create Dataframe

```python
df = pd.DataFrame(record_list)
df
```

```python
def parse_exp_name(x):
    x =  x.split("_")
    d = {'LR':x[2],
         'RD':x[4],
         'EP':x[6]}
    return d

parameters         = pd.DataFrame(list(df['root - INFO - SARSA'].apply(parse_exp_name)))
parameters.columns = ['Learning_Rate', 'Reward_Decay', 'Exploitation_Probability']
parameters         = parameters.astype(float)
```

```python
df
```

```python
gb = pd.concat([df.iloc[:,1:].astype(int), parameters], axis = 1)
```

```python
gb
```

# Worst Case Analysis

```python
def agg_results(gb,col):
    """
    returns gb worst case results
    """
    return gb[col + ['Won','Lost','Draw']].groupby(col).agg(np.max)
```

## Univariate

### Learning Rate

```python
agg_results(gb,['Learning_Rate']).astype(int).Lost.plot(figsize = (20,5))
```

### Reward_Decay (Discount)

```python
agg_results(gb,['Reward_Decay']).astype(int).Lost.plot(figsize = (20,5))
```

### Exploitation_Probability

```python
agg_results(gb,['Exploitation_Probability']).astype(int).Lost.plot(figsize = (20,5))
```

## BiVariate

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Learning_Rate and Reward_Decay

```python
temp = gb.groupby(['Learning_Rate','Reward_Decay']).agg({'Lost':max}).reset_index()
temp_pivot = temp[['Learning_Rate','Reward_Decay','Lost']].pivot('Learning_Rate','Reward_Decay','Lost')

plt.title('Number of Games Lost')
sns.heatmap(temp_pivot, annot=True, fmt="d", cmap="YlGnBu")
```

```python
temp = gb.groupby(['Learning_Rate','Exploitation_Probability']).agg({'Lost':max}).reset_index()
temp_pivot = temp[['Learning_Rate','Exploitation_Probability','Lost']].pivot('Learning_Rate','Exploitation_Probability','Lost')

plt.title('Number of Games Lost')
sns.heatmap(temp_pivot, annot=True, fmt="d", cmap="YlGnBu")
```

# Performance Summary (1000 games)

```python
import seaborn as sns
import numpy as np
import pandas as pd

data = gb

def draw_heatmap(*args, **kwargs):
    data = kwargs.pop('data')
    d = data.pivot(index=args[1], columns=args[0], values=args[2])
    sns.heatmap(d, **kwargs)
```

## Games Lost 

```python
fg = sns.FacetGrid(data, col='Exploitation_Probability',col_wrap = 3)
fg.map_dataframe(draw_heatmap,'Reward_Decay','Learning_Rate' , 'Lost',
                 annot=True, cbar=False, fmt="d", cmap="YlGnBu")
fg.fig.subplots_adjust(top=0.9)
fg.fig.suptitle('Number of Games Lost per 1000 games')
fg.savefig('./plots/GamesLost.png')
```

## Games Drawn

```python
fg = sns.FacetGrid(data, col='Exploitation_Probability',col_wrap = 3)
fg.map_dataframe(draw_heatmap,'Reward_Decay','Learning_Rate' , 'Draw',
                 annot=True, cbar=False, fmt="d", cmap="YlGnBu")
fg.fig.subplots_adjust(top=0.9)
fg.fig.suptitle('Number of Games Draw')
```

## Games Won

```python
fg = sns.FacetGrid(data, col='Exploitation_Probability',col_wrap = 3)
fg.map_dataframe(draw_heatmap,'Reward_Decay','Learning_Rate' , 'Won',
                 annot=True, cbar=False, fmt="d", cmap="YlGnBu")
fg.fig.subplots_adjust(top=0.9)
fg.fig.suptitle('Number of Games Won')
```

## Total Rewards

```python
data['Rewards'] = data.apply(lambda x: np.power(x['Won'] - x['Lost'],3), 1).astype(int)
```
